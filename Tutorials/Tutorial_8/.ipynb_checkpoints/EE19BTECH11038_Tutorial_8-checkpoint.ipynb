```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
```

# DATA

```python
n = 100
x = np.arange(n)+1

e = 5*np.sin(x*3*np.pi/n) + np.sin(x*10*np.pi/n)
d = 1
D1 = 0.08
D2 = 0.025
D3 = 0.005
```

# CVX Variable

```python
h = cp.Variable(n)
```

# Objective

```python
u = h - e

u_plus = cp.pos(u)
u_m_plus = cp.pos(-1*u)

phi_fill = 2*cp.square(u_plus) + 30*u_plus
phi_cut = 12*cp.square(u_m_plus) + u_m_plus

C = phi_fill + phi_cut

Objective = cp.Minimize(cp.sum(C))
```

# Constraints

```python
road_slope = cp.diff(h, k=1)
curvature = cp.diff(h, k=2)
third_derivative = cp.diff(h, k=3)

Constraints = [cp.max(cp.abs(road_slope)) <= D1, 
               cp.max(cp.abs(curvature)) <= D2, 
               cp.max(cp.abs(third_derivative)) <= D3]
```

```python
problem = cp.Problem(objective=Objective, constraints=Constraints)
problem.solve(solver=cp.ECOS)
```

# Convexity of $\phi _{fill}$ snd $\phi _{cut}$

```python
r = np.arange(1, 10, 0.1)

plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.title(r'$\phi _{fill}$')
plt.plot(r, 2*np.power(r, 2) + 30*r)

plt.subplot(1, 2, 2)
plt.title(r'$\phi _{cut}$')
plt.plot(r, 12*np.power(r, 2) + r)
```

## Clearly these are convex functions 

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.plot(h.value)
plt.title("h")

plt.subplot(1, 2, 2)
plt.plot(e)
plt.title("e")

plt.figure(figsize=(16, 8))
plt.plot(h.value-e)
plt.title("h-e")
```

```python
print("Optimal cost is {}".format(np.round(problem.value, 3)))
```
