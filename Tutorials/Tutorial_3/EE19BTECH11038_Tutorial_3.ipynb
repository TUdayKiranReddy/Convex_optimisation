## Problem
First we need to figure out control variables in the problem and those are,
1. $b_1$ which is initial bank balance.
2. $\overline{x}$ which is vector $\begin{bmatrix} x_1 & x_2 & . & x_j & . & x_n \end{bmatrix}^T$ and it represents invest made to $P_{tj}$ which is in our control.
3. $\overline{w}$ which which is vector $\begin{bmatrix} w_1 & w_2 & . & w_t & . & w_n \end{bmatrix}^T$ which represents withdrawls at time t.

Now the initial investment is the objective here where we need to minimize.\
$Objective = min\{b_1 + \mathbf{1}^T\overline{x}\}$ 

And constraints are
1. $b_i \ge 0; \forall i \in [1, n]$, Bank balance shoule be non-negative.
2. $b_i \ge w_i \ge 0; \forall i \in [1, n]$, Since we cannot withdraw more than what we have nor negative cash because negative implies we are depositing which is not mentioned in question.
3. $\overline{x} \ge 0$, investment must be non-negative.
4. $\overline{w} + \mathbf{P}\overline{x} \ge \overline{e}$ given constraint.

## Is it a convex optimisation problem?
Problem is convex iff when both objective and constraints are convex.
1. Let $\overline{c}$ = [$b_1$ $\overline{x}$], Then Objective would be $\mathbf{1}^T\overline{c}$, this is effectively a hyperplane in (n+1)-D. Thus it is convex.
2. Constraints 1, 2, 3 are convex sets since they represents a space without any holes in them.
3. Constraints 4, $\overline{w} + \mathbf{P}\overline{x} \ge \overline{e}$ is just an another linear equation geometrically is an hyperplane. Thus indeed it is convex.
**Therefore the given problem is a convex optimisation problem and thereby can be solved it using *cvxpy***.

### Importing required libraries

```python
import cvxpy as cp
import matplotlib.pyplot as plt
```

###  Control variables
1. *w* is $\overline{w}$
2. *x* is $\overline{x}$
3. *b1* is b1 (Initial bank balance)

```python
w = cp.Variable((n, 1))
x = cp.Variable((m, 1))
b1 = cp.Variable((1, 1))
```

### How to obtain $b_t \forall t$  in an optimised way?
The below transition matrix can obtain $b_t \forall t$. This is possible with just matrix transformation.\
Since,
$$
b_t = (1+\rho)b_{t-1} - w_{t-1} \\
\implies b_t =  (1+\rho)^{t-1}b_1 - \sum_{i=0}^{t-1}(1+\rho)^{t-1-i}w_{i} \\
\begin{bmatrix}
b_1 \\
b_2 \\
b_3\\
. \\
. \\
b_t
\end{bmatrix} = \begin{bmatrix}
1 & 0 & 0 & . & . & 0\\
(1+\rho) & -1 & 0 & . & . & 0\\
(1+\rho)^2 & -(1+\rho) & -1 & . & . & 0\\
(1+\rho)^3 & -(1+\rho)^2 & -(1+\rho) & . & . & 0\\
. & . & . & . & . & .\\
(1+\rho)^{t-1} & -(1+\rho)^{t-2} & -(1+\rho)^{t-3} & . & . & -1\\
\end{bmatrix} \begin{bmatrix}
b_1\\
w_1\\
w_2\\
.\\
.\\
w_{t-1}\\
\end{bmatrix}\\
\overline{b} = \mathbf{M}\begin{bmatrix}
b_1\\
w_1\\
w_2\\
.\\
.\\
w_{t-1}\\
\end{bmatrix}
$$

```python
# transformation matrix M from b1 and w
M = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        if i==j:
            if i==0:
                M[i, j] = 1
            else:
                M[i, j] = -1
        elif i > j:
            if j == 0:
                M[i, j] = np.power(1+rho, i)
            else:
                M[i, j] = -1*np.power(1+rho, i-j)
```

```python
inital_investment = b1 + cp.sum(x)
B = M@cp.vstack([b1, w[0:-1]])

Constraints = [
    B >= 0,
    x >= 0,
    w <= B,
    w >= 0,
    w + P@x >= e
]
```

```python
Objective = cp.Minimize(inital_investment)
problem = cp.Problem(objective=Objective, constraints=Constraints)
problem.solve();
```

```python
print("Minimum initial investment is {}".format(np.round(Objective.value, 2)))
print("\nInitial bank balance is {}".format(np.round(b1.value[0, 0], 2)))
print("\nInitial fund investment balance over various types;")
for i in range(x.value.shape[0]):
    print("\t\tType {} == {}".format(i, np.round(x.value[i, 0], 2)))
print("\nWithdrawls over time")
for i in range(w.value.shape[0]):
    print("\t\tTime {} == {}".format(i+1, np.round(w.value[i, 0], 2)))
```

```python
Time = np.arange(36)+1

plt.figure(figsize=(16, 5))

plt.subplot(1, 3, 1)
plt.title("Expense stream")
plt.plot(Time, e)
plt.grid()

plt.subplot(1, 3, 2)
plt.title("Payouts for different types")
payouts = np.multiply(P, x.value.T)
for i in range(payouts.shape[1]):
    plt.plot(Time, payouts[:, i])
plt.grid()

plt.subplot(1, 3, 3)
plt.title("Bank balance")
plt.plot(B.value)
plt.grid()
```
