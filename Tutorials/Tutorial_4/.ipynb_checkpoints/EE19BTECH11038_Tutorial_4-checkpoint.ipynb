```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
```

# Problem analysis
We are trying to approximate a nth-degree polynomial with (n-1)th degree polynomial and our goal is to find optimal coefficients.
## Objective
$$
argmin_{\forall a_i} \left\{\sum_{k=-N}^{N}\left(x^2_k-\sum_{i=0}^{n-1}a_ix^i_k\right)\right\}
$$
Let $\overline{x}$ be data points, $\mathbf{P}$ be a matrix which includes all power of x till (n-1), $\mathbf{P} = \begin{bmatrix}\overline{1} & \overline{x} & \overline{x}^2 & . & . & \overline{x}^{n-1} \end{bmatrix}_{(2N+1)xn}$.\
Then desired polynomial is $f(\overline{x}) = \overline{x}^n$ and estimated polynomial is $\hat{f}(\overline{x}, \overline{a}) = P\overline{a}$.\
Final objective is $\left(\overline{x}^n-P\overline{a}\right)^T\left(\overline{x}^n-P\overline{a}\right)$
## Constraints
$a_{n-1} \ne 0$, If not it would be (n-2)th degree polynomial.\
**This is not a legal constraint for *CVXPY* framework so this is ignored in below implementation.**

# (a)

```python
def find_optimal_coefficients(n, N=10000, interval=(-1, 1)):
    x = np.linspace(interval[0], interval[1], 2*N+1).reshape(-1, 1)
    
    a = cp.Variable((n, 1))
    P = np.array([np.power(x, i) for i in range(n)])[:, :, 0].T
    
    f_hat = P@a
    f = np.power(x, n)
    error = f - f_hat
    cost =  cp.sum_squares(error)
    
    Objective = cp.Minimize(cost)
    
    Constraints = []
    
    
    problem = cp.Problem(objective=Objective, constraints=Constraints)
    problem.solve();
    
    return a, Objective, f, f_hat, x
```

# (b)

# N=10000, n=5

```python
N = 10000

a, Objective, f, f_hat, x = find_optimal_coefficients(5, N=N)

plt.figure(figsize=(10, 8))
plt.title("n=5, N=1000, Normalised cost={}".format(np.round(Objective.value/(2*N+1), 3)))
plt.plot(x, f, label="f(x)")
plt.plot(x, f_hat.value, label="f'(x)")
plt.xlabel('x')
plt.legend()
plt.grid()
```

# N=10000, n=10

```python
a, Objective, f, f_hat, x = find_optimal_coefficients(10, N=N)

plt.figure(figsize=(10, 8))
plt.title("n=5, N=1000, Normalised cost={}".format(np.round(Objective.value/(2*N+1), 3)))
plt.plot(x, f, label="f(x)")
plt.plot(x, f_hat.value, label="f'(x)")
plt.xlabel('x')
plt.legend()
plt.grid()
```

# N=10000, n=20

```python
a, Objective, f, f_hat, x = find_optimal_coefficients(20, N=N)

plt.figure(figsize=(10, 8))
plt.title("n=5, N=1000, Normalised cost={}".format(np.round(Objective.value/(2*N+1), 3)))
plt.plot(x, f, label="f(x)")
plt.plot(x, f_hat.value, label="f'(x)")
plt.xlabel('x')
plt.legend()
plt.grid()
```

# (c)

```python
n = np.arange(3, 31)

optimal_cost = []
for i in n:
    _, Objective, _, _, _ = find_optimal_coefficients(i, N=N)
    optimal_cost.append(Objective.value)
optimal_cost = np.array(optimal_cost)

plt.figure(figsize=(10, 8))
plt.title("Optimal cost")
plt.plot(n, optimal_cost)
plt.plot(n[n==5], optimal_cost[n==5], "r*")
plt.plot(n[n==10], optimal_cost[n==10], "r*")
plt.plot(n[n==20], optimal_cost[n==20], "r*")
plt.xlabel('n')
plt.ylabel('Optimal cost')
plt.grid()
```

# (d)
## N = 20

```python
def find_optimal_coefficients_abs(n, N=10000, interval=(-1, 1)):
    x = np.linspace(interval[0], interval[1], 2*N+1).reshape(-1, 1)
    
    a = cp.Variable((n, 1))
    P = np.array([np.power(x, i) for i in range(n)])[:, :, 0].T
    
    f_hat = P@a
    f = np.power(x, n)
    error = f - f_hat
    cost =  cp.norm(error, 1)
    
    Objective = cp.Minimize(cost)
    
    Constraints = []
    
    
    problem = cp.Problem(objective=Objective, constraints=Constraints)
    problem.solve();
    
    return a, Objective, f, f_hat, x
```

```python
N = 20

a, Objective, f, f_hat, x = find_optimal_coefficients_abs(5, N=N)

plt.figure(figsize=(10, 8))
plt.title("n=5, N=1000, Normalised cost={}".format(np.round(Objective.value/(2*N+1), 3)))
plt.plot(x, f, label="f(x)")
plt.plot(x, f_hat.value, label="f'(x)")
plt.xlabel('x')
plt.legend()
plt.grid()
```

```python
a, Objective, f, f_hat, x = find_optimal_coefficients_abs(10, N=N)

plt.figure(figsize=(10, 8))
plt.title("n=10, N=1000, Normalised cost={}".format(np.round(Objective.value/(2*N+1), 3)))
plt.plot(x, f, label="f(x)")
plt.plot(x, f_hat.value, label="f'(x)")
plt.xlabel('x')
plt.legend()
plt.grid()
```

```python
a, Objective, f, f_hat, x = find_optimal_coefficients_abs(20, N=N)

plt.figure(figsize=(10, 8))
plt.title("n=20, N=1000, Normalised cost={}".format(np.round(Objective.value/(2*N+1), 3)))
plt.plot(x, f, label="f(x)")
plt.plot(x, f_hat.value, label="f'(x)")
plt.xlabel('x')
plt.legend()
plt.grid()
```

```python
n = np.arange(3, 31)

optimal_cost = []
for i in n:
    _, Objective, _, _, _ = find_optimal_coefficients_abs(i, N=N)
    optimal_cost.append(Objective.value)
optimal_cost = np.array(optimal_cost)

plt.figure(figsize=(10, 8))
plt.title("Optimal cost")
plt.plot(n, optimal_cost)
plt.plot(n[n==5], optimal_cost[n==5], "r*")
plt.plot(n[n==10], optimal_cost[n==10], "r*")
plt.plot(n[n==20], optimal_cost[n==20], "r*")
plt.xlabel('n')
plt.ylabel('Optimal cost')
plt.grid()
```
